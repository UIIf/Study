# <center> Модели машинного обучения

Большинство моделей машинного обучения основаны на математических моделях с различными коэффициентами, называемыми гиперпараметрами. Основной задачей при использовании данных моделей является подбор правильных гиперпараметров, дающих наилучший результат при решении задачи.

# <center> Scikit-learn

Для работы с Машинным обучением в Python используется библиотека Scikit-learn.<br>
<b>Библиотека Scikit-learn</b> — самый распространенный выбор для решения задач классического машинного обучения. Она предоставляет широкий выбор алгоритмов обучения с учителем и без учителя. Одно из основных преимуществ библиотеки состоит в том, что она работает на основе нескольких распространенных математических библиотек, и легко интегрирует их друг с другом. Еще одним преимуществом является широкое сообщество и подробная документация.

Библиотека реализует следующие основные методы:

* <b>Линейные: модели</b>, задача которых построить разделяющую (для классификации) или аппроксимирующую (для регрессии) гиперплоскость.<br>
* <b>Деревья решений</b>: обучение моделей, базирующихся на множестве условий, оптимально выбранных для решения задачи.<br>
* <b>Ансамблевые методы</b>: методы, основанные на деревьях решений, которые комбинируют мощь множества деревьев, и таким образом повышают их качество работы, а также позволяют производить отбор признаков (бустинг, бэггинг, случайный лес, мажоритарное голосование).<br>
* <b>Нейронные сети</b>: комплексный нелинейный метод для задач регрессии и классификации.<br>
* <b>SVM</b>: нелинейный метод, который обучается определять границы принятия решений.<br>
* <b>Наивный Байес</b>: прямое вероятностное моделирование для задач классификации.<br>
* <b>PCA</b>: линейный метод понижения размерности и отбора признаков
* <b>t-SNE</b>: нелинейный метод понижения размерности.<br>
* <b>K-средних</b>: самый распространенный метод для кластеризации, требущий на вход число кластеров, по которым должны быть распределены данные.<br>

# <center> Работа с признаками

Работа с признаками делится на 3 этапа:
1. Извлечение признаков – превращение данных, специфических для предметной области, в понятные для модели векторы
2. Преобразования признаков – трансформация данных для повышения точности алгоритма
3. Выбор признаков – отсечение ненужных признаков

## 1. Извлечение признаков

Признак (feature) - это индивидуальное свойство или характеристика наблюдаемого объекта, которое можно измерить.

### a) Тексты

```python
vectorizer = CountVectorizer()

text = [
     'раз два три',
     'три четыре два два',
     'раз раз раз четыре',
]
X = vectorizer.fit_transform(text)
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

```python
vectorizer = CountVectorizer(ngram_range=(1, 2))

text = [
     'раз два три',
     'три четыре два два',
     'раз раз раз четыре',
]
X = vectorizer.fit_transform(text)
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

<b>TF-IDF</b> (term frequency - inverse document frequency) — статистическая мера, используемая для оценки важности слова в контексте документа, являющегося частью коллекции документов или корпуса. Вес некоторого слова пропорционален частоте употребления этого слова в документе и обратно пропорционален частоте употребления слова во всех документах коллекции.

<b>Word2vec</b> - общее название для совокупности моделей на основе искусственных нейронных сетей, предназначенных для получения векторных представлений слов на естественном языке.

## 2. Преобразования признаков (Feature transformations)

```python
df_num = df[['LatWGS84', 'LonWGS84', 'BottomHoleLatitude',
       'BottomHoleLongitude', 'LATERAL_LENGTH_BLEND', 'PROP_PER_FOOT',
       'WATER_PER_FOOT']]
df_num.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()  # (x - x.mean()) / x.std()   
df_scaled = scaler.fit_transform(df_num)
print(df_scaled[:4])

scaler = MinMaxScaler()  # (x - x.min()) / (x.max() - x.min())  x∈[0,1]
df_scaled = scaler.fit_transform(df_num)
print(df_scaled[:4])
print()
```

```python
# Изменение распределения
from scipy.stats import lognorm

# Содаем случайные данные с логнормальным распределением
data = lognorm(s=1).rvs(1000)
```

```python
plt.hist(data, bins=100)
plt.show()
```

```python
# С помощью логарифмирования приводим к нормальному распределению
plt.hist(np.log(data), bins=100)
plt.show()
```

# <center> Разбиение данных 

При обучении моделей данные разбиваются как правило на три части:<br>
1. Обучающая выборка (<b>train</b>)
2. Тестовая выборка (<b>test</b>)
3. Валидационная выборка (<b>validation</b>) - опционально


Разбиение происходит в соотношении в пользу обучающей выборки:<b>70/15/15%</b>, <b>80/10/10%</b>, либо, без использования валидационной выборки, <b>70/30%</b>, <b>80/20%</b>, <b>90/10%</b>.
    

Помимо этого, используется такой подход как <b>кросс-валидация</b> (cross-validation):

```python
df_main = pd.read_csv('data/wells_info_with_prod.csv')
df_main.head()
```

```python
X = df_main.drop(['Prod1Year', 'ProdAll'], axis=1)
X.head()
```

```python
Y = df_main['ProdAll']
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2, random_state=1)
print(X_train.shape, X_val.shape, X_test.shape)
print(Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from sklearn.model_selection import KFold

CV = KFold(10)
for train_ind, test_ind in CV.split(X): 
    train_X, test_X = X.iloc[train_ind], X.iloc[test_ind]
    train_Y, test_Y = Y.iloc[train_ind], Y.iloc[test_ind]    
```
